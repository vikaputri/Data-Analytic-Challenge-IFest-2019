Pra-proses data(data cleaning)

```
kota= [DKIk,jabark,jatengk,DIYk,jatimk,acehk,sumutk,sumbark,riauk,sumselk,lampungk,kaltengk,kalselk,kaltimk,sulutk,sulselk,sultengk,malukuk,balik,NTBk,bantenk,kepriauk,sulbark,kalutk]
for i in kota :
    i.rename(columns={'Unnamed: 1':'KODE', 'Unnamed: 2':'KABUPATEN/KOTA', 'Unnamed: 3':'JUMLAH SATUAN PENDIDIKAN','Unnamed: 4':'JUMLAH PESERTA','Unnamed: 5':'BAHASA INDONESIA', 'Unnamed: 6':'BAHASA INGGRIS', 'Unnamed: 7':'MATEMATIKA', 'Unnamed: 8':'FISIKA','Unnamed: 9':'KIMIA', 'Unnamed: 10':'BIOLOGI', 'Unnamed: 11':'PKN', 'Unnamed: 12':'RERATA NILAI'},inplace=True)
    i.dropna(inplace = True)
    i.drop(['CAPAIAN NILAI UJIAN NASIONAL'],axis=1)
k = pd.concat(kota)
k = k.drop(['CAPAIAN NILAI UJIAN NASIONAL'],axis=1)
k = k.sort_values('KODE',ascending = True)
k = k.reset_index()
k = k.drop(['index'],axis=1)
k.head(3)
```
